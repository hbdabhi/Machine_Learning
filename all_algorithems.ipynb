```python
data1=pd.read_csv('iris.csv')
data1.head(5)
```

```python
x=data1[['sepal.length','sepal.width','petal.length','petal.width']]
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=3)
ac.fit(x)
```

```python
labels=ac.labels_
labels
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(x)
```

```python
pre=km.predict(x)
pre
```

```python
data1['cluster']=pre
```

```python
data1.head()
```

```python
import matplotlib.pyplot as plt
```

```python
d1=data1[data1.cluster==0]
d2=data1[data1.cluster==1]
d3=data1[data1.cluster==2]
plt.scatter(d1['sepal.length'],d1['sepal.width'],color='black')
plt.scatter(d2['sepal.length'],d2['sepal.width'],color='blue')
plt.scatter(d3['sepal.length'],d3['sepal.width'],color='green')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red')
plt.show()
```

```python
d1=data1[data1.cluster==0]
d2=data1[data1.cluster==1]
d3=data1[data1.cluster==2]
plt.scatter(d1['sepal.length'],d1['sepal.width'],color='black')
plt.scatter(d2['sepal.length'],d2['sepal.width'],color='blue')
plt.scatter(d3['sepal.length'],d3['sepal.width'],color='green')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red',marker="*")
plt.show()
```
